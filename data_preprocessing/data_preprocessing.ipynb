```python
import pandas as pd
```

```python
# 원본 CSV 파일 경로
input_file = "지진대피소(원본).csv"

# 결과 CSV 파일 경로
output_file = "지진대피소.csv"

# CSV 파일 읽기
df = pd.read_csv(input_file, encoding="euc-kr")

# 필요한 컬럼 선택
columns_to_keep = [
    "시설일련번호", "시설명", "시설면적", "경도", "위도", "상세주소", 
    "도로명주소", "관리기관명", "관리부서", "지자체담당자 연락처", 
    "지진겸용 임시주거시실 상세시설명", "최대수용인원수"
]
filtered_df = df[columns_to_keep]

# 새로운 CSV 파일로 저장
filtered_df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"필터링된 데이터가 '{output_file}'에 저장되었습니다.")
```

```python
# CSV 데이터 읽기
df = pd.read_csv("지진대피소.csv", encoding="euc-kr")  # 파일 이름 변경

# 1. 시설정보 테이블 생성
facility_df = df[['시설명', '최대수용인원수', '경도', '위도', '상세주소', '도로명주소']].copy()
facility_df['시설일련번호'] = range(1, len(facility_df) + 1)

# 관리담당자ID와 상세시설ID를 외래키로 추가
facility_df['관리담당자ID'] = range(1, len(facility_df) + 1)
facility_df['상세시설ID'] = range(1, len(facility_df) + 1)

# 2. 관리담당자 테이블 생성
manager_df = df[['담당자', '담당자연락처', '관리기관명']].copy()
manager_df = manager_df.drop_duplicates().reset_index(drop=True)
manager_df['관리담당자ID'] = range(1, len(manager_df) + 1)

# 3. 상세시설정보 테이블 생성
detailed_facility_df = df[['상세시설명']].copy()
detailed_facility_df = detailed_facility_df.drop_duplicates().reset_index(drop=True)
detailed_facility_df['상세시설ID'] = range(1, len(detailed_facility_df) + 1)

# 4. 외래키 연결
# 관리담당자ID 연결
facility_df = facility_df.merge(manager_df[['담당자', '관리담당자ID']], on='담당자', how='left')

# 상세시설ID 연결
facility_df = facility_df.merge(detailed_facility_df[['상세시설명', '상세시설ID']], on='상세시설명', how='left')

# 최종 테이블에서 불필요한 컬럼 제거
facility_df = facility_df[['시설일련번호', '시설명', '최대수용인원수', '경도', '위도', '상세주소', '도로명주소', '관리담당자ID', '상세시설ID']]

# 5. CSV 파일 저장
facility_df.to_csv("시설정보.csv", index=False, encoding="utf-8-sig")
manager_df.to_csv("관리담당자.csv", index=False, encoding="utf-8-sig")
detailed_facility_df.to_csv("상세시설정보.csv", index=False, encoding="utf-8-sig")

print("CSV 파일 저장 완료!")
```
